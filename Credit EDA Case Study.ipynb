### Handling missing values for all the columns

```python
100*(df.isna().sum()/len(df)).round(5)
```

```python
df = df.loc[:, 100*(df.isna().sum()/len(df)).round(5) < .5]
df
```

#### Handling columns whose missing value percentage is around 13%

```python
df.loc[:, 100*(df.isna().sum()/len(df)).round(5) > .13]
```

### Checking the sign of datatypes of all columns
